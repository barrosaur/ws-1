### First formal web scraping project
From .csv file sites given and generated by chatgpt

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

prog_lang = pd.read_csv('programming_languages_sites.csv')
links = prog_lang["site link"]
```

```python
# link 1
link1 = links[0]
response = requests.get(link1).text
soup = BeautifulSoup(response, 'html.parser')
table_ratings = soup.find('table')
headings = table_ratings.find_all('th')

for i, heading in enumerate(headings):
  if heading.text.strip() == "Programming Language":
    idx = i
    
    tbody = table_ratings.find('tbody')
    rows = tbody.find_all('tr')

    for row in rows:
      cols = row.find_all('td')

      # issue: idx is 3 and the third td in a row is an image
      if len(cols) > 4:
        prog_lang = cols[4].text.strip()
        # print(prog_lang)
```
